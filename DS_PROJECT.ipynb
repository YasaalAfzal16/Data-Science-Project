```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

#reading file
df=pd.read_csv('C://Users//M.LAPTOP//Desktop//ForestFire//FW_Veg_Rem_Combined.csv')
df.drop(df.columns[[0, 1]], axis = 1, inplace = True)

```

```python
df.info()
```

```python
#converting stat_cause_descr, state & discovery_month to int

df['stat_cause_descr']=pd.factorize(df['stat_cause_descr'])[0]

df['state']=pd.factorize(df['state'])[0]

df['discovery_month']=pd.factorize(df['discovery_month'])[0]
```

```python
df.isnull().sum()
```

```python
df.remoteness=df.remoteness.fillna(df.remoteness.mode)
```

```python
'''
Discretizing:
b/w 0 to 0.14    : nearest ->1
b/w 0.14 to 0.20 : near ->2
b/w 0.20 to 0.30 : far ->3
b/w 0.30 to 1.00 : farthest->4
'''

df["remoteness"]=pd.cut(df["remoteness"],
       bins=[0, 0.14, 0.20, 0.30, 1.00], 
       labels=[1, 2, 3, 4])
```

```python
# remoteness:
df["remoteness"].value_counts().plot(kind='pie')
plt.show()
```

```python
#Important Step
df['putout_time']=df['putout_time'].str.split("days").str[0] # num __[days]__ timer --> num | __[days]__ timer
df['putout_time']
```

```python
#converting to numeric
df['putout_time']=pd.to_numeric(df['putout_time'], errors='coerce')
```

```python
#converting to int with mean inserted
df['putout_time']=df['putout_time'].fillna(df['putout_time'].mean()).astype('int')
```

```python
df=df.rename(columns={'putout_time': 'putout_time(days)'})
```

```python
#dropping unnecessary attributes:
df=df.drop(['fire_size'], axis=1)
```

```python
#dropping unnecessary attributes:
df=df.drop(['fire_size_class'], axis=1)
```

```python
'''
Discretizing:
0.1 : low ->1
1.0 || 3.0 || 10.0: medium ->2
50.0 ||100.0 : high ->3 
'''
df['fire_mag']=df['fire_mag'].map({0.1:1, 1.0:2, 3.0:2, 10.0:2, 50.0:3, 100.0:3}).astype(int)
```

```python
fire_magnitude after discretizing:
df['fire_mag'].value_counts().plot(kind='pie')
plt.show()
```

```python
#cause of fire:
df['stat_cause_descr'].value_counts().plot(kind='pie')
plt.show()
```

```python
'''
Discretizing:
0 || 4 || 9 : low ->1
12 || 14 : medium ->2
15 || 16 : high ->3 
'''
df['Vegetation']=df['Vegetation'].map({0:1, 4:1, 9:1,
                                       12:2, 14:2,
                                       15:3, 16:3}).astype(int)
```

```python
#Vegetation after discretizing:
df['Vegetation'].value_counts().plot(kind='pie')
plt.show()
```

```python
# ____DECISION-TREE____

from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
```

```python
#splitting the dataset in an 80:20 ratio for training and test set respectively:

X=df.drop(columns=['fire_mag']).values
Y=df['fire_mag'].values
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
#Letâ€™s take a look at the shape of the created train and test sets:

print('Shape of X_train=>',X_train.shape)
print('Shape of X_test=>',X_test.shape)
print('Shape of Y_train=>',Y_train.shape)
print('Shape of Y_test=>',Y_test.shape)
```

```python
df.remoteness=df.remoteness.fillna(df.remoteness.mean())
```

```python
# Building Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion = 'gini', random_state = 42)
dt.fit(X_train, Y_train)
dt_pred_train = dt.predict(X_train)
```

```python
# Evaluation on Training set
dt_pred_train = dt.predict(X_train)
print('Training Set Evaluation F1-Score(micro)=>',f1_score(Y_train,dt_pred_train ,average='micro'))
```

```python
# Evaluating on Test set
dt_pred_test = dt.predict(X_test)
print('Testing Set Evaluation F1-Score=>(micro)',f1_score(Y_test,dt_pred_test, average='micro'))
```

```python
# ___RANDOM-FOREST____

# Building  Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(criterion = 'entropy', random_state = 42)
rfc.fit(X_train, Y_train)

# Evaluating on Training set
rfc_pred_train = rfc.predict(X_train)
print('Training Set Evaluation F1-Score=>',f1_score(Y_train,rfc_pred_train, average='micro'))
```

```python
# Evaluating on Test set
rfc_pred_test = rfc.predict(X_test)
print('Testing Set Evaluation F1-Score=>',f1_score(Y_test,rfc_pred_test, average='micro'))
```

```python
# Here, we can clearly see that the random forest model performed much better than the decision tree in the out-of-sample
# evaluation
```

```python
import numpy as np
feature_importance=pd.DataFrame({
    'rfc':rfc.feature_importances_,
    'dt':dt.feature_importances_
},index=df.drop(columns=['fire_mag']).columns)
feature_importance.sort_values(by='rfc',ascending=True,inplace=True)

index = np.arange(len(feature_importance))
fig, ax = plt.subplots(figsize=(18,8))
rfc_feature=ax.barh(index,feature_importance['rfc'],0.4,color='purple',label='Random Forest')
dt_feature=ax.barh(index+0.4,feature_importance['dt'],0.4,color='lightgreen',label='Decision Tree')
ax.set(yticks=index+0.4,yticklabels=feature_importance.index)

ax.legend()
plt.show()
```
